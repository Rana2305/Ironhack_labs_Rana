## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
customer_amount_spent = orders.groupby('CustomerID',as_index=False).agg({'amount_spent':'sum'})
customer_amount_spent
```

```python
VIP_Customers  = customer_amount_spent.loc[customer_amount_spent['amount_spent']>=customer_amount_spent. amount_spent.quantile(0.95)]
VIP_Customers 
```

```python
Preferred_Customers= customer_amount_spent.loc[(customer_amount_spent['amount_spent']>=customer_amount_spent. amount_spent.quantile(0.75))& (customer_amount_spent['amount_spent']<customer_amount_spent. amount_spent.quantile(0.95))]
Preferred_Customers
```

Now we'll leave it to you to solve Q2 & Q3, which you can leverage from your solution for Q1:

## Q2: How to identify which country has the most VIP Customers?

```python
country_VIP = pd.merge(orders,VIP_Customers, on = 'CustomerID').groupby('Country',as_index=False).agg({'CustomerID':'nunique'})

```

```python
country_VIP.sort_values('CustomerID', ascending = False)
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
VIP_preferred_customers_country = pd.merge(Preferred_Customers, orders , on ='CustomerID').groupby('Country',as_index=False).agg({'CustomerID':'nunique'})
VIP_preferred_customers_country.sort_values
```

```python
VIP_preferred_customers_country.sort_values('CustomerID', ascending = False)
```
